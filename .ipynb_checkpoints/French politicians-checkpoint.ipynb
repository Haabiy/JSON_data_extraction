### List of French politicians

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
from time import sleep
from enum import Enum
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
url = 'https://fr.wikipedia.org/wiki/Liste_des_communaut%C3%A9s_d%27agglom%C3%A9ration_par_r%C3%A9gion'
driver.get(url)
sleep(3)
```

```python
store_links = []
soup = BeautifulSoup(driver.page_source, 'html.parser')
for div in soup.find_all('a', {'class':'sidebar-toc-link'}):
    if div.text:
        print(div['href'])
        store_links.append(div['href'])
```

```python
store_links.pop(0)
```

```python
for x in store_links:
    print(x)
```

```python
for i in range(len(store_links)):
    print(i, ".", url+store_links[i])
```

```python
# https://medium.com/analytics-vidhya/web-scraping-a-wikipedia-table-into-a-dataframe-c52617e1f451
def func(wikiurl):
    driver.get(wikiurl)
    soup_ = BeautifulSoup(driver.page_source, 'html.parser')
    table = soup_.find('table', {'class' : 'wikitable alternance sortable jquery-tablesorter'})
    df=pd.read_html(str(table))
    # convert list to dataframe
    df=pd.DataFrame(df[0])
    return df
```

```python
func(url+store_links[3])
```
